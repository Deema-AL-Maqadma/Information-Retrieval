```python
# في هذه الخلية نقوم بتحميل جميع الملفات النصية الموجودة داخل مجلد documents
# الهدف: قراءة كل ملف .txt وإضافة محتواه إلى قائمة documents لاستخدامه لاحقًا في المعالجة وبناء نموذج TF-IDF

import os

path = "/content/documents"   # مسار المجلد الذي يحتوي على ملفات النصوص
documents = []                # قائمة لتخزين محتوى كل مستند
filenames = []

# قراءة الملفات النصية داخل المجلد بالترتيب
for filename in sorted(os.listdir(path)):
    with open(os.path.join(path, filename), "r", encoding="utf-8") as f:
        documents.append(f.read())   # إضافة محتوى الملف إلى القائمة
        filenames.append(filename)
```

```python
# هذه الخلية للتأكد من أن المجلد تم رفعه بشكل صحيح
# وتعرض أسماء الملفات الموجودة داخل مجلد documents

os.listdir("/content/documents")
```

```python
# في هذه الخلية نقوم بعرض قائمة كلمات الـ Stopwords
# الهدف: معرفة عدد الكلمات التي سيتم حذفها أثناء عملية تنظيف النصوص
# هذه الكلمات تعتبر شائعة جداً ولا تضيف معنى مهم لعملية البحث

from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# طباعة عدد كلمات الـ stopwords
print("عدد كلمات الـ Stopwords:", len(list(ENGLISH_STOP_WORDS)))

# طباعة قائمة كاملة بالـ stopwords
print("قائمة الـ Stopwords:")
print(list(ENGLISH_STOP_WORDS))
```

```python
# في هذه الخلية نقوم بعمل الـ preprocessing للنصوص
# الخطوات:
# 1) تحويل النص إلى lowercase
# 2) إزالة علامات الترقيم
# 3) إزالة كلمات الـ stopwords
# 4) إعادة تجميع الكلمات في نص نظيف

import string
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

cleaned_docs = []   # قائمة لحفظ النصوص بعد التنظيف

# جدول لإزالة علامات الترقيم
punct_table = str.maketrans("", "", string.punctuation)

for doc in documents:
    text = doc.lower()                     # تحويل إلى lowercase
    text = text.translate(punct_table)     # إزالة علامات الترقيم
    words = text.split()                   # تقسيم النص إلى كلمات
    words = [w for w in words if w not in ENGLISH_STOP_WORDS]  # إزالة stopwords
    cleaned_docs.append(" ".join(words))   # إعادة تجميع النص

cleaned_docs[:3]   # عرض أول 3 مستندات بعد التنظيف
```

```python
# في هذه الخلية نقوم ببناء نموذج TF-IDF
# هذا النموذج يحول النصوص إلى تمثيل عددي يعتمد على أهمية الكلمات

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()          # إنشاء النموذج
tfidf_matrix = vectorizer.fit_transform(cleaned_docs)   # تدريب النموذج على المستندات

tfidf_matrix    # عرض المصفوفة الناتجة
```

```python
# في هذه الخلية نعرض قائمة الكلمات (vocabulary)
# وهي الكلمات التي اعتمد عليها نموذج TF-IDF

vocabulary = vectorizer.get_feature_names_out()
vocabulary
```

```python
# في هذه الخلية نحول مصفوفة TF-IDF إلى جدول (DataFrame)
# لعرض القيم بشكل أوضح

import pandas as pd

tfidf_dense = tfidf_matrix.toarray()               # تحويل إلى مصفوفة كثيفة
tfidf_df = pd.DataFrame(tfidf_dense, columns=vocabulary)

tfidf_df.head()    # عرض أول 5 صفوف
```

```python
# في هذه الخلية ندخل استعلام المستخدم
# ونطبق عليه نفس خطوات الـ preprocessing

query = input("Enter your search query: ")

# تنظيف الاستعلام
q = query.lower()
q = q.translate(punct_table)
q_words = q.split()
q_words = [w for w in q_words if w not in ENGLISH_STOP_WORDS]
cleaned_query = " ".join(q_words)

# تحويل الاستعلام إلى متجه TF-IDF
query_vec = vectorizer.transform([cleaned_query])

query_vec
```

```python
# في هذه الخلية نحسب درجة التشابه بين الاستعلام وكل المستندات
# باستخدام Cosine Similarity

from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(query_vec, tfidf_matrix)
similarities
```

```python
# في هذه الخلية نرتب المستندات حسب أعلى درجة تشابه
# ثم نعرض النتائج من الأعلى إلى الأقل

import numpy as np

ranking = np.argsort(similarities[0])[::-1]   # ترتيب تنازلي

for idx in ranking:
    print(f"Score: {similarities[0][idx]:.4f} | Document: {filenames[idx]}")
```
